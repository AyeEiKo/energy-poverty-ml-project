```python
# This notebook focuses on feature engineering and missing value imputation
import numpy as np
import pandas as pd

df = pd.read_parquet("MYANMAR/DATA/2.df_MM_MEPI_Score_Energy_Poor.parquet", engine="pyarrow")
```

```python
print(df["EnergyPoor"].value_counts(dropna=False))
print("dtype:", df["EnergyPoor"].dtype)
```

```python
#conver to int
df["EnergyPoor"] = df["EnergyPoor"].astype("int8")
```

```python
print("Describe:")
print(df["hv220AgeOfHead"].describe())

print("\nMissing count:")
print(df["hv220AgeOfHead"].isna().sum())

print("\nDtype:")
print(df["hv220AgeOfHead"].dtype)
```

```python
print("Describe:")
print(df["hv009FamilySize"].describe())

print("\nMissing count:")
print(df["hv009FamilySize"].isna().sum())

print("\nDtype:")
print(df["hv009FamilySize"].dtype)
```

```python
print("Describe:")
print(df["hv216HouseSize"].describe())

print("\nMissing count:")
print(df["hv216HouseSize"].isna().sum())

print("\nDtype:")
print(df["hv216HouseSize"].dtype)
```

```python
#run to impute
df["hv216HouseSize"] = df["hv216HouseSize"].fillna(
    df["hv216HouseSize"].median()
)
```

```python
print("Describe:")
print(df["hv014NoOfChildren"].describe())

print("\nMissing count:")
print(df["hv014NoOfChildren"].isna().sum())

print("\nDtype:")
print(df["hv014NoOfChildren"].dtype)
```

```python
df["hv247HasBankAccount"] = df["hv247HasBankAccount"].astype("category")
```

```python
print("Value counts:")
print(df["v714CurrentlyWorking"].value_counts(dropna=False))

print("\nDtype:")
print(df["v714CurrentlyWorking"].dtype)
```

```python
print("Value counts (top 10):")
print(df["hv115_01MaritalStatus"].value_counts(dropna=False).head(10))

print("\nDtype:")
print(df["hv115_01MaritalStatus"].dtype)
```

```python
df["hv115_01MaritalStatus"] = df["hv115_01MaritalStatus"].astype("category")
df["hv115_01MaritalStatus"] = (
    df["hv115_01MaritalStatus"]
    .cat.add_categories("Missing")
    .fillna("Missing")
)
```

```python
print("Value counts:")
print(df["v717Occupation"].value_counts(dropna=False))

print("\nDtype:")
print(df["v717Occupation"].dtype)
```

```python
df["v717Occupation"] = df["v717Occupation"].astype("category")
```

```python
print("Value counts:")
print(df["745aHouseOwnership"].value_counts(dropna=False))

print("\nDtype:")
print(df["745aHouseOwnership"].dtype)
```

```python
df["745aHouseOwnership"] = df["745aHouseOwnership"].astype("category")
```

```python
df["hhidCaseIdentification"] = df["hhidCaseIdentification"].astype("string")
df["hv000CountryCode"] = df["hv000CountryCode"].astype("category")
df["hv001ClusterNumber"] = df["hv001ClusterNumber"].astype("int32")
df["hv002HouseholdNumber"] = df["hv002HouseholdNumber"].astype("int32")
```

```python
print("Remaining columns:")
print(df.columns.tolist())

print("\nDtypes:")
print(df.dtypes)

print("\nMissing values:")
print(df.isna().sum().sort_values(ascending=False).head())
```

```python
cols_to_string = [
    "hv115_01MaritalStatus",
]
df[cols_to_string] = df[cols_to_string].astype("string")
```

```python
df.to_parquet("MYANMAR/DATA/3.df_MM_FeatureClean.parquet",engine="pyarrow",compression="snappy"
)

print(f"✅ File saved.")
```
