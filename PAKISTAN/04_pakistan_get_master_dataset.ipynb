```python
# This notebook constructs the final master dataset by integrating feature engineering outputs and target variables for modelling
import numpy as np
import pandas as pd

df_feature = pd.read_parquet("PAKISTAN/DATA/3.df_PK_FeatureClean.parquet", engine="pyarrow")
```

```python
print("\nDtypes:")
print(df_feature.dtypes)
```

```python
print("\nDtypes:")
print(df_feature.dtypes)

print("\nMissing values:")
print(df_feature.isna().sum().sort_values(ascending=False).head())
```

```python
print("\nDtypes:")
print(df_feature.dtypes)

print("\nMissing values:")
print(df_feature.isna().sum().sort_values(ascending=False).head())
```

```python
#normalize to numeric first
cols_to_numeric = [
    "hv024RegionDivision",
    "hv025TypeOfResidence",
    "hv219SexOfHead",
    "hv270WealthIndex",  
    "hv106_01EducationOfHead",
    "hv115_01MaritalStatus",
    "v717Occupation",
    "745aHouseOwnership",
]

for col in cols_to_numeric:
    df_feature[col] = pd.to_numeric(
        df_feature[col],
        errors="coerce"
    )
```

```python
def restore_ml_dtypes(df):
    df = df.copy()

    df["hv220AgeOfHead"] = df["hv220AgeOfHead"].astype("int8")
    df["hv009FamilySize"] = df["hv009FamilySize"].astype("int8")
    df["hv216HouseSize"] = df["hv216HouseSize"].astype("float64")
    df["hv014NoOfChildren"] = df["hv014NoOfChildren"].astype("int8")

    df[[
        "hv024RegionDivision",
        "hv025TypeOfResidence",
        "hv219SexOfHead",
        "hv270WealthIndex",
        "hv106_01EducationOfHead",
        "hv115_01MaritalStatus",
        "v717Occupation",
        "745aHouseOwnership",
        "hv247HasBankAccount",
        "v714CurrentlyWorking",
    ]] = df[[
        "hv024RegionDivision",
        "hv025TypeOfResidence",
        "hv219SexOfHead",
        "hv270WealthIndex",
        "hv106_01EducationOfHead",
        "hv115_01MaritalStatus",
        "v717Occupation",
        "745aHouseOwnership",
        "hv247HasBankAccount",
        "v714CurrentlyWorking",
    ]].astype("category")

    df[[
        "EnergyPoor"
    ]] = df[[
        "EnergyPoor"
    ]].astype("int8")

    return df
```

```python
df_master = restore_ml_dtypes(df_feature)
```

```python
print("\nDtypes:")
print(df_master.dtypes)

print("\nMissing values:")
print(df_master.isna().sum().sort_values(ascending=False).head())
```

```python
df_master.to_parquet("PAKISTAN/DATA/4.PK_MASTER_DATASET.parquet",engine="pyarrow", compression="snappy", index=False)
print(f"✅ File saved.")
```
